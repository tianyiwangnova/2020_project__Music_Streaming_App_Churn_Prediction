```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
pd.set_option('display.max_columns', 500)
```

```python
data = pd.read_csv("features table.csv")
```

```python
plt.rcParams['figure.figsize'] = (20, 20)
style.use('Solarize_Light2')

plt.subplot(5,4,1)
plt.hist(data['active_days'])
plt.title('Active days')

plt.subplot(5,4,2)
plt.hist(data['sessions_per_day'])
plt.title('Average numbers of sessions per day')

plt.subplot(5,4,3)
plt.hist(data['time_per_session']/(1000*60*60))
plt.title('Average hours per session')

plt.subplot(5,4,4)
plt.hist(data['time_per_day']/(1000*60*60))
plt.title('Average hours per day')

plt.subplot(5,4,5)
plt.hist(data['avg_items_per_session'])
plt.title('Average numbers of items per session')

plt.subplot(5,4,6)
plt.hist(data['unique_songs'])
plt.title('Numbers of unique songs')

plt.subplot(5,4,7)
plt.hist(data['unique_artists'])
plt.title('Numbers of unique artists')

plt.subplot(5,4,8)
plt.hist(data['unique_artists'])
plt.title('Numbers of unique artists')

plt.subplot(5,4,9)
plt.hist(data['perc_1_6'])
plt.title('1am ~ 6am percentage')

plt.subplot(5,4,10)
plt.hist(data['perc_6_12'])
plt.title('6am ~ 12pm percentage')

plt.subplot(5,4,11)
plt.hist(data['perc_12_18'])
plt.title('12pm ~ 18pm percentage')

plt.subplot(5,4,12)
plt.hist(data['perc_18_1'])
plt.title('18pm ~ 1am percentage')

plt.subplot(5,4,13)
plt.hist(data['perc_addlist'])
plt.title('Percentage of addtolist')

plt.subplot(5,4,14)
plt.hist(data['perc_thdn'])
plt.title('Percenrage of thumbsdown')

plt.subplot(5,4,15)
plt.hist(data['perc_thup'])
plt.title('Percentage of thumbsup')

plt.subplot(5,4,16)
plt.hist(data['perc_popular_songs'])
plt.title('Percentage of popular songs')

plt.subplot(5,4,17)
plt.hist(data['perc_paid_days'])
plt.title('Percentage of paid days')

plt.subplot(5,4,18)
plt.hist(data['female'])
plt.title('Female users')

plt.subplot(5,4,19)
plt.hist(data['last_status_paid'])
plt.title('Last status paid')

plt.subplot(5,4,20)
plt.hist(data['customer_age'])
plt.title('Customer age')
```

```python
print("{:.2%} of the customers spent more than 3 hours per active day".format(len(data[data['time_per_day']/(1000*60*60) > 3])/len(data)))
```

```python
print("{:.2%} of the customers remained paid status for the whole time period".format(len(data[data['perc_paid_days'] == 1])/len(data)))
```
